```python
#Createing a dataframe with all reviews values
reviews = listings[["id" , "price", "room_type" , "review_scores_checkin" , "review_scores_cleanliness" , "review_scores_communication" ,
                     "review_scores_location" , "review_scores_rating" , "review_scores_value"]]
reviews = reviews.rename(columns = {"review_scores_checkin" : "Check-in" , "review_scores_cleanliness" : "Cleanliness" ,
                                    "review_scores_communication" : "Communication" , "review_scores_location" : "Location" ,
                                    "review_scores_rating" : "Rating" , "review_scores_value" : "Value"})

#Modifying the range of "Ratings" from 0 - 100 to 1-10
reviews["Rating"] = reviews["Rating"]/10

#Creating "Mean" to visualize wich rows are all NaN reviews values
reviews_NaN = reviews.copy()
reviews_NaN["Mean"] = reviews[["Check-in", "Cleanliness", "Communication", "Location", "Rating" ,"Value"]].mean(axis=1).round(1)

print(reviews_NaN.isnull().sum())
```

```python
#Reviews' statistical descriptions per room type 
reviews["room_type"] = listings["room_type"] 
EntireReviews = reviews[reviews["room_type"] == "Entire home/apt"]
PrivateReviews  = reviews[reviews["room_type"] == "Private room"]
SharedReviews = reviews[reviews["room_type"] == "Shared room"]

EntireReviews.describe()
```

```python
PrivateReviews.describe()
```

```python
SharedReviews.describe()
```

```python
#Createing a melt version of reviews
reviews_id = reviews.melt(id_vars = ["id" , "price", "room_type"])
reviews_id = reviews_id.rename(columns = {"variable" : "Review" , "value" : "Score"})

reviewsall = reviews_id[["room_type", "Score" , "Review" , "price"]]

reviewsall["Score"] = round(reviewsall["Score"])
reviewsall = reviewsall.pivot_table(index = "Score", columns = "Review", values = "price")
reviewsall.fillna(0)

reviewsRT = reviews_id[["room_type" , "Score" , "Review" , "price"]]
reviewsRT["Score"] = round(reviewsRT["Score"])

EntiereHome = reviewsRT[reviewsRT["room_type"] == "Entire home/apt"]
EntiereHome = EntiereHome.pivot_table(index = "Score", columns = "Review", values = "price")
EntiereHome.fillna(0)

PrivateRoom = reviewsRT[reviewsRT["room_type"] == "Private room"]
PrivateRoom = PrivateRoom.pivot_table(index = "Score", columns = "Review", values = "price")
PrivateRoom.fillna(0)

SharedRoom = reviewsRT[reviewsRT["room_type"] == "Shared room"]
SharedRoom = SharedRoom.pivot_table(index = "Score", columns = "Review", values = "price")
SharedRoom.fillna(0)

f, ax = plt.subplots(figsize=(16, 8))
a = sns.heatmap(reviewsall, annot=True, linewidths=0.5, ax=ax , cmap = "flare").set_title("All Room Types")

f, ax = plt.subplots(figsize=(16, 8))
b = sns.heatmap(EntiereHome, annot=True, linewidths=0.5, ax=ax , cmap = "flare").set_title("Entiere Home/Apt")

f, ax = plt.subplots(figsize=(16, 8))
c = sns.heatmap(PrivateRoom, annot=True, linewidths=0.5, ax=ax , cmap = "flare").set_title("Private Room")

f, ax = plt.subplots(figsize=(16, 8))
d = sns.heatmap(SharedRoom, annot=True, linewidths=0.5, ax=ax , cmap = "flare").set_title("Shared Room")
```

```python
#Creating df with the amenities
listings["amenities"] = listings["amenities"].astype(str).str.title()
amenities = listings.amenities.str.replace('\{|"|\}',"").str.get_dummies(',')

#Adding id from listings dataframe
amenities[["id"]] = listings[["id"]]
colnames = amenities.columns.tolist()
colnames = colnames[-1:] + colnames[:-1]
amenities = amenities[colnames]

amenities
```

```python
amenities["all"] = amenities.iloc[:, 1:-1].sum(axis=1)

amenities["all"].describe()
```

```python
#Melted version of amenities
amenities_id = amenities.melt(id_vars = ["id"])
amenities_id = amenities_id.rename(columns = {"variable" : "Amenitie" , "value" : "Freq"})

#amenities df grouped by amenitie
amenities_grouped = amenities_id.groupby("Amenitie").sum().reset_index().sort_values("Freq" , ascending = False)
amenities_grouped.tail(20)
```
